```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
with open('New_All_Pokemon.txt', 'r') as file:
    data = file.read().replace('\n', '')
soup = BeautifulSoup(data, 'html.parser')
newSoup = soup.prettify()

indexBeg = newSoup.find('table')
indexEnd = newSoup.find('</table >')

dfList = pd.read_html(newSoup[indexBeg:indexEnd])
dftable = dfList[0]
dftable
```

```python
string_array = []
packs = dftable["Set Name"].unique()
for name in packs:
    string_array.append(str(name))
string_array.sort()
string_array
```

```python
pack_lists = []
for pack_name in string_array:
    pack_df = dftable.loc[dftable["Set Name"] == pack_name]
    pack_lists.append(pack_df)
pack_lists[0]
```
